```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv", index_col="Id")
df
```

```python
df.info()
```

```python
object_categories = []
for col in list(df.columns):
    if str(df[col].dtype) == "object":
        print(f"{col} category has {len(df[col].unique())} categories")
        object_categories.append(col)
```

```python
df[object_categories]
```

```python
numeral_colummns = []
for col in list(df.columns):
    if str(df[col].dtype) in ["int64", "float64"]:
        numeral_colummns.append(col)
```

```python
df[numeral_colummns]
```

```python
num_zeros = {}
for column in numeral_colummns:
    num_zeros[column] = sum(df[column] == 0)
```

```python
sorted(num_zeros.items(), key=lambda x:x[1])
```
